```python
import pandas as pd

products = pd.read_csv("../data/raw/products.csv")
behavior = pd.read_csv("../data/raw/user_behavior.csv")

products.head()
behavior.head()
```

```python
products.info()
behavior.info()

products.isnull().sum()
behavior.isnull().sum()
```

```python
products['category'] = products['category'].str.capitalize()
products['material'] = products['material'].str.capitalize()
products['color'] = products['color'].str.capitalize()
products['style'] = products['style'].str.capitalize()
```

```python
# price → fill with median
products['price'] = products['price'].fillna(products['price'].median())

# rating → fill with mean
products['rating'] = products['rating'].fillna(products['rating'].mean())
```

```python
products = products.drop_duplicates()
behavior = behavior.drop_duplicates()
```

```python
products = products[products['price'] > 0]
products = products[products['rating'] <= 5]
```

```python
products.to_parquet("../data/processed/products_clean.parquet", index=False)
behavior.to_parquet("../data/processed/user_behavior_clean.parquet", index=False)
```
